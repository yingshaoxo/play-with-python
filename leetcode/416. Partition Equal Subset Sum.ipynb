```python
from itertools import permutations

class Solution:
    def canPartition(self, nums: List[int]) -> bool:
        if sum(nums) % 2 != 0:
            return False
        
        nums_x = permutations(nums, len(nums))

        for nums in nums_x:
            for i in range(len(nums)):
                if i != 0:
                    a = list(nums[i:])
                    b = list(nums[:i])
                    if sum(a) == sum(b):
                        return True
        return False
```
