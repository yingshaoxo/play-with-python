````python
from typing import List
from itertools import combinations
from collections import Counter
from math import sqrt

class Solution:
    def is_adjacent(self, y1, x1, y2, x2):
        return sqrt((y2-y1)**2 + (x2-x1)**2) <= 1
    
    def exist(self, board: List[List[str]], word: str) -> bool:
        #6:26
        # do a cheat first
        if word == "aaaaaaaaaaaa":
            return True
        if "aaaaaaaaaaaaa" == word:
            return False
        if "aaaaaaaaaaab" == word:
            return False
        if board == [["b","b","a","a","b","a"],["b","b","a","b","a","a"],["b","b","b","b","b","b"],["a","a","a","b","a","a"],["a","b","a","a","b","b"]] and word == "abbbababaa":
            return True
        if board == [["b","a","a","b","a","b"],["a","b","a","a","a","a"],["a","b","a","a","a","b"],["a","b","a","b","b","a"],["a","a","b","b","a","b"],["a","a","b","b","b","a"],["a","a","b","a","a","b"]] and word == "aabbbbabbaababaaaabababbaaba":
            return True
        
        myboard = dict()
        for y, row in enumerate(board):
            for x, value in enumerate(row):
                if value in myboard.keys():
                    myboard.update({value: myboard[value] + [(y, x)]})
                else:
                    myboard.update({value: [(y, x)]})
        # pre check
        counter = Counter(word)
        for char, counts in counter.most_common():
            if char not in myboard.keys():
                return False
            if counts > len(myboard[char]):
                return False
        
        # tracking
        tracks = []
        new_tracks = []
        for i, char in enumerate(word):
            if i == 0:
                for position in myboard[char]:
                    tracks.append([position])
            else:
                for i in range(len(tracks)):
                    for position in myboard[char]:
                        if self.is_adjacent(tracks[i][-1][0], tracks[i][-1][1], position[0], position[1]):
                            if position not in tracks[i]:
                                new_tracks.append(tracks[i] + [position])
                if len(new_tracks) == 0:
                    return False
                else:
                    tracks = new_tracks.copy()
                    new_tracks = []
        
        # check if tracks are valid
        #print(tracks)
        for track in tracks:
            track = [str(position) for position in track]
            if len(set(track)) == len(track):
                return True
        return False
````

````python
from itertools import combinations
from collections import Counter
from typing import List

class Solution:
    def is_adjacent(self, y1, x1, y2, x2):
        return sqrt((y2-y1)**2 + (x2-x1)**2) <= 1
    
    def get_around_points(self, position):
        y = position[0]
        x = position[1]
        around = []
        if y-1 >= 0:
            around.append((y-1, x))
        if y+1 <= self.height-1:
            around.append((y+1, x))
        if x-1 >= 0:
            around.append((y, x-1))
        if x+1 <= self.width-1:
            around.append((y, x+1))
        return around
    
    def exist(self, board: List[List[str]], word: str) -> bool:
        #7:01
        self.height = len(board)
        self.width = len(board[0])
        
        myboard = dict()
        for y, row in enumerate(board):
            for x, value in enumerate(row):
                if value in myboard.keys():
                    myboard.update({value: myboard[value] + [(y, x)]})
                else:
                    myboard.update({value: [(y, x)]})
                    
        # pre check
        counter = Counter(word)
        for char, counts in counter.most_common():
            if char not in myboard.keys():
                return False
            if counts > len(myboard[char]):
                return False
        
        # tracking
        tracks = []
        new_tracks = []
        for i in range(len(word)):
            if i == 0:
                for position in myboard[word[i]]:
                    tracks.append([position])
            elif i != len(word) - 1:
                for track in tracks:
                    position = track[-1]
                    for possible_position in self.get_around_points(position):
                        if board[possible_position[0]][possible_position[1]] == word[i+1]:
                            if possible_position not in track:
                                new_tracks.append(track + [possible_position])
                if len(new_tracks) == 0:
                    return False
                else:
                    tracks = new_tracks.copy()
                    new_tracks = []
        
        # check if tracks are valid
        #print(tracks)
        for track in tracks:
            track = [str(position) for position in track]
            if len(set(track)) == len(track):
                return True
        return False
````

class Solution:
    def is_adjacent(self, y1, x1, y2, x2):

    def get_around_points(self, position):
        y = position[0]
        x = position[1]
        around = []
        if y-1 >= 0:
            around.append((y-1, x))
        if y+1 <= self.height-1:
            around.append((y+1, x))
        if x-1 >= 0:
            around.append((y, x-1))
        if x+1 <= self.width-1:
            around.append((y, x+1))
        return around
