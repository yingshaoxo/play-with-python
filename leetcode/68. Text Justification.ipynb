```python
class Solution:
    def fullJustify(self, words: List[str], maxWidth: int) -> List[str]:
        #2022.1.16 8:07
        lines = []
        
        oneLine = ""
        oneLineWords = []
        for i, word in enumerate(words):
            if i != 0:
                oneLine += " "
            tempSentence = oneLine + word
            tempSentence = tempSentence.strip()
            if (len(tempSentence) > maxWidth):
                theLine = ""
                totallyChar = 0
                for w in oneLineWords:
                    totallyChar += len(w)
                if len(oneLineWords) == 1:
                    oneLine = oneLineWords[0].ljust(maxWidth, " ")
                    lines.append(oneLine)
                    
                    oneLine = ""
                    oneLineWords = []

                    oneLine += word
                    oneLineWords.append(word)
                else:
                    slotNumber = len(oneLineWords) - 1
                    dividor = (maxWidth - totallyChar)/slotNumber
                    if isinstance(dividor, float):
                        perfectDividor = " " * ((maxWidth - totallyChar)//slotNumber)
                        spaceRequired = maxWidth - totallyChar
                        slotsList = [perfectDividor for i in range(slotNumber)]
                        for index, w in enumerate(range(slotNumber)):
                            if len("".join(slotsList)) == spaceRequired:
                                break
                            slotsList[index] = slotsList[index] + " "
                        for w in oneLineWords:
                            theLine += w 
                            if len(slotsList):
                                theLine += slotsList[0]
                                slotsList = slotsList[1:]
                    else:
                        theLine = (dividor*" ").join(oneLineWords)
                    lines.append(theLine)
                    oneLine = ""
                    oneLineWords = []

                    oneLine += word
                    oneLineWords.append(word)
                
            elif (len(tempSentence) == maxWidth):
                theLine = oneLine + word
                lines.append(theLine)
                oneLine = ""
                oneLineWords = []
            else:
                oneLine += word
                oneLineWords.append(word)
                
        lines = [line.lstrip() for line in lines]
        if oneLine != "":
            oneLine = oneLine.lstrip()
            oneLine = oneLine.ljust(maxWidth, " ")
            lines.append(oneLine)
        
        return lines
        #2022.1.6 8:27
        #debug until 9:11
```
